```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df=pd.read_csv("/content/Women_s_E-Commerce_Clothing_Reviews_1594_1.csv",sep=';')
df.head(4)
```

```python
x=df['Review.Text']
y=df['Rating']
print(x.shape,y.shape)
```

```python
import nltk
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
stwd=stopwords.words('english')
lem=WordNetLemmatizer()
x=np.array(x)
```

```python
word_tokenize(x[0])
```

```python
x_sample=[]
for sent in x:

  words=word_tokenize(str(sent))
  words=[(lem.lemmatize(word)).lower() for word in words if word not in stwd]
  x_sample.append(words)
x_sample
```

```python
print(len(x_sample),y.shape)
```

```python
!pip install gensim
!pip install python-Levenshtein
```

```python
import gensim
from gensim.models import Word2Vec
model=Word2Vec(window=5,min_count=2,workers=4)
model.build_vocab(x_sample,progress_per=1000)
```

```python
model.train(x_sample,total_examples=len(x_sample),epochs=5)
```

```python
model.save("/content/word2vec.h5")
```

```python
model.wv.most_similar('favorite')
```
